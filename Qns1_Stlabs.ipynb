```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt('/content/Data_1.txt')
data2 = np.loadtxt('/content/Data_2.txt')
```

```python
def find_max_min(data):
    max = []
    min = []

    for i in range(1, len(data) - 1):
        if data[i-1] < data[i] > data[i+1]:
            max.append(i)
        if data[i-1] > data[i] < data[i+1]:
            min.append(i)
    return max, min
peaks1, valleys1 = find_max_min(data1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data1, label='Data-1')
plt.plot(peaks1, data1[peaks1], 'ro')
plt.plot(valleys1, data1[valleys1], 'bo')
plt.title('Data-1')
plt.show()
```

```python
print("Max for Data_1: ", peaks1)
print("Min for Data_1: ", valleys1)
```

```python
peaks2, valleys2 = find_max_min(data2)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data2, label='Data-2')
plt.plot(peaks2, data2[peaks2], 'ro')
plt.plot(valleys2, data2[valleys2], 'bo')
plt.title('Data-2')
plt.show()
```

```python
print("Max for Data_2: ", peaks2)
print("Min for Data_2: ", valleys2)
```
